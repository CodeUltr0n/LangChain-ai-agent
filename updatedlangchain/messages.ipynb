### Messages

Messages are how you talk to the model. Each message has a role that tells the model who is speaking.                             
   
  Types                                                                                                                             
                  
  HumanMessage — You (the user)

  from langchain_core.messages import HumanMessage                                                                                  
  HumanMessage(content="What is the weather in India?")
                                                                                                                                    
  AIMessage — The model (response)                                                                                                  
                                                                                                                                    
  from langchain_core.messages import AIMessage                                                                                     
  AIMessage(content="The weather in India is sunny.")                                                                               
  When the model wants to call a tool, it returns an AIMessage with tool_calls instead of content.                                  
                                                                                                                                    
  SystemMessage — Instructions for the model                                                                                        
                                                                                                                                    
  from langchain_core.messages import SystemMessage
  SystemMessage(content="You are a helpful weather assistant.")                                                                     
  Tells the model how to behave. Set once at the beginning.                                                                         
                                                                                                                                    
  ToolMessage — Result from a tool                                                                                                  
                                                                                                                                    
  from langchain_core.messages import ToolMessage
  ToolMessage(content="It's sunny in Gujarat", tool_call_id="abc123")                                                               
  The result you send back after running a tool. Must include the tool_call_id so the model knows which tool call this result       
  belongs to.                                                                                                                       
                                                                                                                                    
  Flow                                                                                                                              
                                                                                                                                    
  HumanMessage → Model → AIMessage (with tool_calls)                                                                                
                            ↓                                                                                                       
                      You run the tool                                                                                              
                            ↓                                                                                                       
  ToolMessage → Model → AIMessage (final answer)                                                                                    
                                                                                                                                    
  Think of it like a conversation thread — each message is a turn, and the model reads the full thread to understand context.

```python
import os 
from langchain.chat_models import init_chat_model
from dotenv import load_dotenv

load_dotenv()
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
model = init_chat_model("groq:llama-3.3-70b-versatile")
```

```python
model.invoke("please write a short note on e-commerce business in 2026" ) # simple human message --> text prompts
```

```python
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage 

messages = [
    SystemMessage(content="You are a poetry expert"),
    HumanMessage(content="Write a poem on purity of earth")
]

response = model.invoke(messages)
print(response.content)
```

```python
# See the model thinking - Stream (word by word generation)
messages = [
    SystemMessage(content="You are a poetry expert"),
    HumanMessage(content="Write a poem on purity of earth")
]

print("Model is thinking and generating...\n")
for chunk in model.stream(messages):
    print(chunk.content, end="", flush=True)
```

```python
# Debug mode - shows token usage, model info, and response details
from langchain_core.messages import SystemMessage, HumanMessage

messages = [
    SystemMessage(content="You are a poetry expert"),
    HumanMessage(content="Write a short poem on purity of earth")
]

response = model.invoke(messages)

# Show the full reasoning details
print("=== Model Thinking Details ===")
print(f"Model: {response.response_metadata.get('model_name', 'unknown')}")
print(f"Finish reason: {response.response_metadata.get('finish_reason', 'unknown')}")
print(f"System fingerprint: {response.response_metadata.get('system_fingerprint', 'N/A')}")
print(f"\n=== Token Breakdown ===")
token_usage = response.response_metadata.get('token_usage', {})
print(f"Input tokens: {token_usage.get('prompt_tokens', 0)}")
print(f"Output tokens: {token_usage.get('completion_tokens', 0)}")
print(f"Total tokens: {token_usage.get('total_tokens', 0)}")
print(f"Time taken: {token_usage.get('total_time', 0):.3f}s")
print(f"\n=== Final Response ===")
print(response.content)
```

```python
# See agent reasoning with tool calls
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_core.messages import HumanMessage

@tool
def get_weather(location: str) -> str:
    """Get the current weather at a location"""
    return f"It's sunny in {location}, 25°C"

# Use init_chat_model with proper Groq config
agent_model = init_chat_model("groq:llama-3.3-70b-versatile", temperature=0)

agent = create_agent(
    model=agent_model,
    tools=[get_weather]
)

print("Agent thinking process:\n")

# Use invoke to see the full reasoning chain
response = agent.invoke({"messages": [HumanMessage(content="what is the weather in germany?")]})

for msg in response["messages"]:
    if msg.type == "ai" and hasattr(msg, 'tool_calls') and msg.tool_calls:
        print(f"[AI Thinking] I need to call: {msg.tool_calls[0]['name']}({msg.tool_calls[0]['args']})")
    elif msg.type == "tool":
        print(f"[Tool Result] {msg.content}")
    elif msg.type == "ai" and msg.content:
        print(f"[AI Response] {msg.content}")
```

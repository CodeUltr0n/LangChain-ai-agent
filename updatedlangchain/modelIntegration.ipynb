### streaming and batch

 Streaming in LLMs means receiving the response token by token as it's generated, instead of waiting for the entire response to        
  complete. When you call model.stream(), the model starts sending back small chunks of text in real-time — each chunk is a few tokens  
  that the model has just predicted. This gives you a live, word-by-word output effect, similar to how you see ChatGPT typing in its UI.
   It feels faster because you see the first token immediately, rather than waiting 5-10 seconds for the full 200-word response to
  finish before seeing anything.
  

```python
for chunk in model.stream("write me a 200 word paragraph on world war 2"):
    print(chunk.text,end="",flush=True)
```

### Batch in LLMs means sending multiple prompts at once and getting all responses back together after they're all done. When you call ─
  model.batch([prompt1, prompt2, prompt3]), the model processes all inputs in parallel (depending on rate limits) and returns a list  
  of complete responses. It's efficient when you have many prompts to run — instead of calling the model 10 times sequentially, you   
  send all 10 together and wait once. The tradeoff is you wait longer upfront but get better overall throughput. 

```python
model.batch(
    ["why do parrots talk?",
    "how do planes fly?",
    "what is quantum computing?"],
    config={
        "max_concurrency":5, #limit to 5 parallel calls
    }
)
```
